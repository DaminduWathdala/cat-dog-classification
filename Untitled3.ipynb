```python
data.drop_duplicates(inplace=True)
```

```python
data["Hours Studied"] = data["Hours Studied"].astype(float)
```

```python

data["Previous Scores"] = data["Previous Scores"].astype(float)
data["Sleep Hours"] = data["Sleep Hours"].astype(float)
data["Sample Question Papers Practiced"] = data["Sample Question Papers Practiced"].astype(float)
```

```python
data
```

```python
data = pd.get_dummies(data, columns=["Extracurricular Activities"])
```

```python
data
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
data["Performance Index"] = scaler.fit_transform(data[["Performance Index"]])
data["Previous Scores"] = scaler.fit_transform(data[["Previous Scores"]])
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(data["sleep Hours"],data["Performance Index"])
plt.show()
```

```python
group1 = data.groupby("Hours Studied")["Performance Index"].mean()
```

```python
group2 = data.groupby("Previous Scores")["Performance Index"].mean()
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.subplot(2,2,1)
plt.plot(group1.index,group1.values)
plt.xlabel("Hours Studied")
plt.ylabel("Performance Index")

plt.subplot(2,2,2)
plt.plot(group2.index,group2.values)
plt.xlabel("Previous Scores")
plt.ylabel("Performance Index")

group3 = data.groupby("Sleep Hours")["Performance Index"].mean()
group4 = data.groupby("Sample Question Papers Practiced")["Performance Index"].mean()

plt.subplot(2,2,3)
plt.plot(group3.index,group3.values)
plt.xlabel("Sleep Hours")
plt.ylabel("Performance Index")

plt.subplot(2,2,4)
plt.plot(group4.index,group4.values)
plt.xlabel("Sample Question Papers Practiced")
plt.ylabel("Performance Index")

plt.show()
```
