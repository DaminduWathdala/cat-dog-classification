```python
import numpy as np
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],np.power([1,2,3,4],2),'ks:')
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
plt.plot(x,np.power(x,2),'ks:', x,np.power(x,3))
plt.show()
```

```python
plt.plot([1,2,3,4],[5,6,7,8])
left,right = plt.xlim(2,4)
bottom ,top = plt.ylim(6,8)
print(left,right)
print(bottom,top)
plt.show()
```
